```python
import plotly.io as pio
import plotly.express as px
import pandas as pd
import numpy as np

df = pd.read_csv('data/import/BankChurners.csv')
df.head(5)



```

```python
df.insert(1, 'Attrition_Num', df['Attrition_Flag'].apply(lambda x: 1 if x == 'Existing Customer' else 0))
df.head(5)
```

#Coding attrition.

```python
fig = px.scatter(df, x="Customer_Age", y="Income_Category", color="Attrition_Flag")
fig.show()
```

#It shows that mainly the age 36-59 customers are churning, regardless their income.

```python
fig = px.scatter_3d(df, 
                    x="Marital_Status", 
                    y="Months_on_book", 
                    z="Card_Category",
                    color='Attrition_Flag')

fig.update_layout(title="Marital Status, Months on Book and Card Category by Attrition Status")

fig.show()       
```

#Blue card owners are the most loyal customers.

```python
import plotly.express as px

fig = px.histogram(df, x="Customer_Age")
fig.show()
```

```python
fig = px.parallel_categories(df, 
                             dimensions=["Attrition_Flag", "Gender", "Marital_Status", "Card_Category"])

fig.update_layout(title="Parallel Categories: Attrition Status, Gender, Marital Status, and Card Category")

fig.show()
```

```python
fig = px.parallel_categories(
    df,
    dimensions=["Attrition_Flag", "Gender", "Marital_Status", "Card_Category"],
    color=df["Attrition_Flag"].astype("category").cat.codes,  # convert to numeric for coloring
    color_continuous_scale=px.colors.sequential.Viridis
)

fig.update_layout(title="Parallel Categories: Customers by Attrition Status and Attributes")

fig.show()
```
